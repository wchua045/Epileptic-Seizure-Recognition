```python
#create new DF for instances where y==1
Class1=df.loc[df.y == 1]
Class1
```

```python
#create new DF for instances where y==2
Class2=df.loc[df.y == 2]
Class2
```

```python
#create new DF for instances where y==4
Class4=df.loc[df.y == 4]
Class4
```

```python
#create new DF for instances where y==5
Class5=df.loc[df.y == 5]
Class5
```

```python
#calculate mean EEG for every timestamp and form new DF containing mean values
Class1mean=Class1.mean(axis = 0) 
Class1mean
```

```python
#calculate mean EEG for every timestamp and form new DF containing mean values
Class2mean=Class2.mean(axis = 0) 
Class2mean
```

```python
#calculate mean EEG for every timestamp and form new DF containing mean values
Class3mean=Class3.mean(axis = 0) 
Class3mean
```

```python
#calculate mean EEG for every timestamp and form new DF containing mean values
Class4mean=Class4.mean(axis = 0)  
Class4mean
```

```python
#calculate mean EEG for every timestamp and form new DF containing mean values
Class5mean=Class5.mean(axis = 0) 
Class5mean
```

```python
#concat all the new DF containing mean values for Class 1,2,3,4,5
mergedDF=pd.concat([Class1mean,Class2mean,Class3mean,Class4mean,Class5mean], axis=1)
#rename column name
mergedDF.columns=['Class 1','Class 2','Class 3','Class 4','Class 5']
#drop 'y' row
newDF=mergedDF.drop('y')
newDF
```

```python
# Create traces
#Plot line graph using mean values for every timestamp, for the different classes
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=newDF.index, y=newDF['Class 1'],mode='lines',name='Class 1'))
fig.add_trace(go.Scatter(x=newDF.index, y=newDF['Class 2'],mode='lines',name='Class 2'))
fig.add_trace(go.Scatter(x=newDF.index, y=newDF['Class 3'],mode='lines',name='Class 3'))
fig.add_trace(go.Scatter(x=newDF.index, y=newDF['Class 4'],mode='lines',name='Class 4'))
fig.add_trace(go.Scatter(x=newDF.index, y=newDF['Class 5'],mode='lines',name='Class 5'))
fig.show()


```

```python
#concat all the DF containing standard deviation for Class 1,2,3,4,5
std_dev_DF=pd.concat([Class1.std(axis=0),Class2.std(axis=0),Class3.std(axis=0),Class4.std(axis=0),Class5.std(axis=0)], axis=1)
#rename column name
std_dev_DF.columns=['Class 1 std dev','Class 2 std dev','Class 3 std dev','Class 4 std dev','Class 5 std dev']
#drop 'y' row
std_dev_DF=std_dev_DF.drop('y')
std_dev_DF

```

```python
# Create traces
#Plot line graph using standard deviation for every timestamp, for the different classes
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=std_dev_DF.index, y=std_dev_DF['Class 1 std dev'],mode='lines',name='Class 1 std dev'))
fig.add_trace(go.Scatter(x=std_dev_DF.index, y=std_dev_DF['Class 2 std dev'],mode='lines',name='Class 2 std dev'))
fig.add_trace(go.Scatter(x=std_dev_DF.index, y=std_dev_DF['Class 3 std dev'],mode='lines',name='Class 3 std dev'))
fig.add_trace(go.Scatter(x=std_dev_DF.index, y=std_dev_DF['Class 4 std dev'],mode='lines',name='Class 4 std dev'))
fig.add_trace(go.Scatter(x=std_dev_DF.index, y=std_dev_DF['Class 5 std dev'],mode='lines',name='Class 5 std dev'))
fig.show()


```

```python
#concat all the DF containing maximum values for Class 1,2,3,4,5
max_DF=pd.concat([Class1.max(),Class2.max(),Class3.max(),Class4.max(),Class5.max()], axis=1)
#rename column name
max_DF.columns=['Class 1 max','Class 2 max','Class 3 max','Class 4 max','Class 5 max']
#drop 'y' row
max_DF=max_DF.drop('y')
max_DF
```

```python
# Create traces
#Plot line graph using maximum value for every timestamp, for the different classes
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=max_DF.index, y=max_DF['Class 1 max'],mode='lines',name='Class 1 max'))
fig.add_trace(go.Scatter(x=max_DF.index, y=max_DF['Class 2 max'],mode='lines',name='Class 2 max'))
fig.add_trace(go.Scatter(x=max_DF.index, y=max_DF['Class 3 max'],mode='lines',name='Class 3 max'))
fig.add_trace(go.Scatter(x=max_DF.index, y=max_DF['Class 4 max'],mode='lines',name='Class 4 max'))
fig.add_trace(go.Scatter(x=max_DF.index, y=max_DF['Class 5 max'],mode='lines',name='Class 5 max'))
fig.show()
```
